## FEATURES

The various feature of the cleaned dataset are explained below:

1.AIRLINE: The name of the airline company is stored in the airline column.It is a categorical feature having 6 different airlines.
2.FLIGHT : Flight store information regarding the plane flight code. it is a categorical feature.
3.SOURCE CITY : City from which the flight takes off. it is a categorical feature having 6 unique cities.
4.DEPARTURE TIME: This is a derived categorical feature obtained created by grouping time periods into bins. it stores information about the 
   departure time and have 6 unique time lables.
5.STOPS: Acategorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6.ARRIVAL TIME: This is a derived categorical feature created by grouping time intervals into bins. its has six distinct time labels and keeps information
about the arrival time.
7.DESTINATION CITY: City where the flight will land.it is a categorical feature having 6 unique cities.
8.CLASS: A continuous feature that contains information on seat class; it has two distinct values; Business and Economy.
9.DURATION: A continous feature that displays the overall amount of time it takes to travel between cities in hours.
10.DAYS LEFT:This is a derived characteristics that is calculated by subtracting the trip date by the booking date.
11.PRICE: Target variable stores information of the ticket price.

```python
## Feature Engineering 
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop Date of Journey

df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```
